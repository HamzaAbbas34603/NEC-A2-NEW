```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVR, SVC
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
```

```python
# Load the Boston dataset
# Loading and Preprocessing the Dataset
boston_data_path = 'boston_house_prices.csv'
boston = pd.read_csv(boston_data_path, header=1)
 
boston
```

```python
# Assuming the target variable is in a column named 'MEDV'
X_boston = boston.drop('MEDV', axis=1)
y_boston = boston['MEDV']

# Split the dataset into training and test sets
X_train_boston, X_test_boston, y_train_boston, y_test_boston = train_test_split(X_boston, y_boston, test_size=0.2, random_state=42)
```

```python

# SVM for regression
svm_regressor = SVR(kernel='rbf', C=1, gamma='scale')
svm_regressor.fit(X_train_boston, y_train_boston)

# Predict on the test set for regression
svm_pred_boston = svm_regressor.predict(X_test_boston)

threshold = 25
# SVM for binary classification
svm_binary = SVC(kernel='rbf', C=1, gamma='scale', probability=True, random_state=0)
svm_binary.fit(X_train_boston, (y_train_boston > threshold).astype(int))

# Predict on the test set for binary classification
svm_pred_binary_boston = svm_binary.predict(X_test_boston)
svm_pred_proba_boston = svm_binary.predict_proba(X_test_boston)[:, 1]

# Compute Accuracy, AUC, Confusion Matrix, and Classification Error for binary classification
accuracy_svm = accuracy_score((y_test_boston > threshold).astype(int), svm_pred_binary_boston)
conf_matrix_svm = confusion_matrix((y_test_boston > threshold).astype(int), svm_pred_binary_boston)
roc_auc_svm = roc_auc_score((y_test_boston > threshold).astype(int), svm_pred_proba_boston)
classification_error_svm = 1 - accuracy_svm

# Print metrics for SVM (binary classification)
print("\nMetrics for SVM (Binary Classification) on Test Set:")
print("Accuracy:", accuracy_svm)
print("AUC:", roc_auc_svm)
print("Confusion Matrix:")
print(conf_matrix_svm)
print("Classification Error:", classification_error_svm)
```
